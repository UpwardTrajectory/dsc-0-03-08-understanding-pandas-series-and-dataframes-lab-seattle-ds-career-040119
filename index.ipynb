## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

```python
#df.columns
station_names = df[['linename','station']].drop_duplicates()
count_stops = station_names['station'].value_counts()

def get_count_stops(x):
    return count_stops[x]

df['Num_Lines'] = df['station'].map(lambda x:get_count_stops(x))
df = df.sort_values(by=['Num_Lines'], ascending=False)
df.head()
```
